# ðŸ“Š Hands-on Exploratory Data Analysis (EDA)

**EDA** is the initial step in data analysis, aimed at understanding the characteristics, structure, and patterns within a dataset before applying statistical modeling or machine learning.

EDA is typically carried out through:

- **Descriptive summaries** â†’ simple statistics such as mean, median, standard deviation, and frequency counts.

- **Visualization** â†’ plots (histograms, scatter plots, boxplots, heatmaps, etc.) to uncover patterns, relationships between variables, and anomalies.

```python
# Mengekstrak variabel numerik dari URL

# Panjang total URL
data['url_length'] = data['url'].str.len()

# Panjang domain (hostname)
from urllib.parse import urlparse
data['domain_length'] = data['url'].apply(lambda x: len(urlparse(x).netloc))

# Panjang path (setelah domain)
data['path_length'] = data['url'].apply(lambda x: len(urlparse(x).path))

# Jumlah parameter query (?key=value&...)
data['num_params'] = data['url'].apply(lambda x: urlparse(x).query.count('&') + (1 if urlparse(x).query else 0))

# Jumlah titik (.)
data['num_dots'] = data['url'].str.count(r'\.')

# Jumlah dash (-)
data['num_hyphens'] = data['url'].str.count(r'-')

# Jumlah angka (0â€“9)
data['num_digits'] = data['url'].str.count(r'\d')

# Ada https atau tidak
data['is_https'] = data['url'].apply(lambda x: 1 if urlparse(x).scheme == "https" else 0)

data.head()
```

```python
# Boxplot untuk deteksi outlier visual

plt.figure(figsize=(6,4))
sns.boxplot(x=data_clean["url_length"])
plt.title("Boxplot Outlier Detection")
plt.show()
```

```python
# visualize range 
# Create a histogram
plt.hist(data_clean["url_length"], bins=30, edgecolor='black', alpha=0.5, color='green')

# Add vertical line
plt.axvline(x=min(data_clean["url_length"]), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(data_clean["url_length"]):.2f}')
plt.axvline(x=max(data_clean["url_length"]), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(data_clean["url_length"]):.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Outlier Mean and Std Method
mean = data_clean["url_length"].mean()
std = data_clean["url_length"].std()

upper_limit = mean + 2*std
lower_limit = mean - 2*std

print(f"Mean: {mean:.2f}, Std: {std:.2f}")
print(f"Upper limit: {upper_limit:.2f}, Lower limit: {lower_limit:.2f}")

outliers = data_clean[(data_clean["url_length"] > upper_limit) | (data_clean["url_length"] < lower_limit)]

print("\nNormal:")
print(len(data_clean) - len(outliers))

print("\nOutliers (Mean Â± 2*Std):")
print(len(outliers))
```

```python
# Buat histogram
plt.hist(data_clean["url_length"], bins=30, edgecolor='black', alpha=0.5, color='green')

# Tambahkan garis vertikal pada standar deviasi
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=lower_limit, color='blue', linestyle='dashed', linewidth=1, label=f'{lower_limit:.2f}')
plt.axvline(x=upper_limit, color='blue', linestyle='dashed', linewidth=1, label=f'{upper_limit:.2f}')

# Tambahkan label dan judul
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Std Deviation = {std:.2f}')

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
```

```python
# Outlier IQR Method
Q1 = data_clean["url_length"].quantile(0.25)
Q2 = data_clean["url_length"].quantile(0.5)
Q3 = data_clean["url_length"].quantile(0.75)
print("Q1:", Q1)
print("Q3:", Q3)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outliers_iqr = data_clean[(data_clean["url_length"] < lower_bound) | (data_clean["url_length"] > upper_bound)]

print("\nNormal:")
print(len(data_clean) - len(outliers_iqr))

print("\nOutliers (IQR):")
print(len(outliers_iqr))
```

```python
# Buat histogram
plt.hist(data_clean["url_length"], bins=30, edgecolor='black', alpha=0.5, color='green')

# Tambahkan garis vertikal pada standar deviasi
plt.axvline(x=Q2, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=lower_bound, color='blue', linestyle='dashed', linewidth=1, label=f'{lower_bound:.2f}')
plt.axvline(x=upper_bound, color='blue', linestyle='dashed', linewidth=1, label=f'{upper_bound:.2f}')

# Tambahkan label dan judul
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'IQR = {IQR:.2f}')

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
```

### Strategi Penanganan Outlier

* **EDA tahap awal** â†’ cukup identifikasi & laporkan outlier.
* **Kalau outlier karena error** â†’ hapus.
* **Kalau outlier bagian dari fenomena** â†’ simpan untuk analisis khusus.
* **Kalau outlier membuat distribusi berat sebelah** â†’ pertimbangkan transformasi (log, sqrt)

```python
# hapus semua outlier
# menggunakan yang mean std method

data_normal = data_clean[
    ((data_clean["url_length"] > lower_limit) & (data_clean["url_length"] < upper_limit)) 
]
data_normal.head()
```

```python
import math 

# 1. Menentukan banyak data
n = data_normal["url_length"].count()
print("1. n =", n)

# membuat histogram secara manual untuk kolom url_length
d_min = data_normal["url_length"].min()
d_max = data_normal["url_length"].max()
print("2. Dmin =", d_min, ", Dmax =", d_max)

# 3. Menghitung rentang data
R = d_max - d_min
print("3. R =", R)

# 4. Menentukan banyak kelas
k = math.ceil(1 + 3.3 * math.log(n, 10))
print("4. Banyak kelas =", k)

# 5. Menghitung panjang interval kelas
I = math.ceil(R / k)
print("5. Panjang interval kelas =", I)

# 6. Menghitung interval kelas dan tepi kelas
kelas = []
interval_kelas = []
tepi_kelas = [] # hanya ambil BAK - 0,5
for i in range(k):
    BAK = d_min + i * I
    BBK = BAK + I
    tepi = BAK - 0.5
    kelas.append(i + 1)
    interval_kelas.append(str(BAK) + "-" + str(BBK))
    tepi_kelas.append(tepi)
tepi_kelas.append(BBK + 0.5)
print("6. Kelas =", kelas)
print("6. Interval kelas =", interval_kelas)
print("6. Tepi kelas =", tepi_kelas)


# 7. Menghitung frekuensi tiap kelas

# Convert data into a pandas DataFrame
df = data_normal.copy()

# Create frequency distribution using pd.cut and value_counts
df["range"] = pd.cut(df["url_length"], bins=tepi_kelas, labels=interval_kelas, include_lowest=True)
frequency_table = df["range"].value_counts().sort_index()
relative_frequency = frequency_table / frequency_table.sum() * 100

# Convert to DataFrame for a cleaner look
frequency_distribution = pd.DataFrame(
    {
        "Kelas": kelas,
        "Range": frequency_table.index,
        "Frequency": frequency_table.values,
        "Relative Frequency (%)": relative_frequency,
    }
)
frequency_distribution.reset_index(drop=True, inplace=True)

# Display the frequency distribution table
print("\nTable of Frequency Distribution")
print(frequency_distribution)
```

```python
# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.bar(
    frequency_distribution["Range"],
    frequency_distribution["Frequency"],
    width=1,
    edgecolor="black",
    color="skyblue",
)
plt.title("Histogram of Frequency Distribution")
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.grid(axis="y")

# Show the plot
plt.tight_layout()
plt.show()
```

### Cara Membaca

* Nilai korelasi berada di rentang **-1 sampai +1**:

  * `+1` â†’ hubungan linear positif sempurna.
  * `-1` â†’ hubungan linear negatif sempurna.
  * `0` â†’ tidak ada hubungan linear.
* Korelasi tinggi (>|0.7|) dapat menandakan potensi **multicollinearity**.

```python
# visualize correlation using heatmaps
sns.heatmap(data_corr.corr(method='pearson'),
            annot = True);

plt.show()
```

* `is_https` â†” `domain_length` = **0.23** â†’ domain panjang agak sering menggunakan HTTPS.

#### Korelasi dengan `type_encoded` (target)
