### 3. Distribusi Probabilitas untuk Variabel Numerik

Menghitung dan memvisualisasikan distribusi untuk variabel `Age` sebagai variabel acak kontinu

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Distribusi Usia
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title("Distribusi Usia Pengguna")
plt.xlabel("Usia")
plt.ylabel("Frekuensi")
plt.show()
```

### 4. Uji Distribusi

Untuk mengetahui jenis distribusi data, kita bisa melakukan dua cara
1. **Analisis visual**
   * Analisis visual bisa menggunakan histogram dan Q-Q plot
   * Jika Q-Q plot menunjukkan bahwa data mengikuti garis lurus, itu bisa menjadi indikasi bahwa data mengikuti distribusi tertentu
3. **Uji statistik**
   * 
   * Hasil dari uji statistik (seperti p-value) akan membantu kita menentukan apakah data cukup kuat untuk menolak hipotesis bahwa data berasal dari distribusi tertentu

```python
import scipy.stats as stats
```

```python
# Membuat Q-Q Plot
plt.figure(figsize=(8, 6))
stats.probplot(df['Age'], dist="norm", plot=plt)
plt.title('Q-Q Plot untuk Memeriksa Normalitas')
plt.xlabel('Teoretis Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid()
plt.show()
```

Hasil: **Tidak normal**

```python
# Membuat Q-Q Plot
plt.figure(figsize=(8, 6))
stats.probplot(df['Age'], dist="uniform", plot=plt)
plt.title('Q-Q Plot untuk Memeriksa Distribusi Uniform')
plt.xlabel('Teoretis Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid()
plt.show()
```

Hasil: **Uniform**

### 6. Distribusi Probabilitas untuk Variabel Kategoris

Memvisualisasikan distribusi untuk variabel kategoris `Location` dan `Platform`

```python
# Menghitung distribusi platform yang digunakan
platform_counts_normalized = df['Platform'].value_counts(normalize=True)
platform_counts = df['Platform'].value_counts()

print("\nDistribusi Probabilitas Platform yang digunakan")
print(platform_counts)

print("\nDistribusi Probabilitas Platform yang digunakan (Normalized)")
print(platform_counts_normalized)
```

Mengembalikan frekuensi relatif dari setiap nilai

```python
# Visualisasi Distribusi Tingkat Kepuasan
sns.countplot(data=df, x='Platform')
plt.title("Distribusi Platform yang digunakan")
plt.xlabel("Platform")
plt.ylabel("Frekuensi")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd

# Contoh data: Ganti dengan data Anda
data = np.random.normal(loc=0, scale=1, size=1000)

# 1. Visualisasi Data
# Histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Histogram')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.show()

# Q-Q Plot
stats.probplot(data, dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.show()

# 2. Statistik Deskriptif
df = pd.DataFrame(data, columns=['Data'])
print(df.describe())

# 3. Uji Kesesuaian Distribusi


# Uji Shapiro-Wilk
shapiro_statistic, p_value_sw = stats.shapiro(data)
print(f"Shapiro-Wilk Statistic: {shapiro_statistic}, P-Value: {p_value_sw}")
```
