## 1. Table of Frequency Distribution

Menyajikan data dengan lebih informatif dalam bentuk tabel.

Langkah-Langkah:
1. Menentukan banyak data `n = 40`
2. Menentukan data minimum dan maksimum `Dmin = 66, Dmax = 144`
3. Menghitung rentang data `R = Dmax – Dmin = 144 – 66 = 78`
4. Menentukan banyak kelas dengan menggunakan **kaidah empiris Sturgess** `k = 1 + 3,3 log(n)` -> `k = 1 + 3,3 log(40) = 1 + 5,29 = 6,29` dibulatkan menjadi `7` (kelas harus bilangan bulat, sehingga selalu dibulatkan ke atas)
5. Menentukan panjang kelas interval `I = R / k -> I = 78 / 7 = 11,2` dibulatkan ke atas menjadi `12`
6. Menentukan tepi kelas (class boundaries) dengan rumus `BBK – 0,5` dan `BAK + 0,5 `
7. Menghitung frekuensi tiap kelas menggunakan `Tally`.
8. Menghitung frekuensi absolut `fi`  dan frekuensi relatif tiap kelas `fr = fi / n * 100%`

```python
import pandas as pd
import math

# Sample data
data = [
    66, 76, 79, 86, 89, 97, 125, 110, 105, 111, 111, 102, 101, 121, 117, 122, 120, 101, 115, 100, 115, 95, 122, 139, 144, 137, 128, 138, 136, 136, 92, 97, 106, 81, 112, 107, 106, 133, 133, 103,
]

# 1. Menentukan banyak data
n = len(data)
print("1. n =", n)

# 2. Menentukan data minimum dan maksimum
d_min = min(data)
d_max = max(data)
print("2. Dmin =", d_min, ", Dmax =", d_max)

# 3. Menghitung rentang data
R = d_max - d_min
print("3. R =", R)

# 4. Menentukan banyak kelas
k = math.ceil(1 + 3.3 * math.log(n, 10))
print("4. Banyak kelas =", k)

# 5. Menghitung panjang interval kelas
I = math.ceil(R / k)
print("5. Panjang interval kelas =", I)

# 6. Menghitung interval kelas dan tepi kelas
kelas = []
interval_kelas = []
tepi_kelas = [] # hanya ambil BAK - 0,5
for i in range(k):
    BAK = d_min + i * I
    BBK = BAK + I
    tepi = BAK - 0.5
    kelas.append(i + 1)
    interval_kelas.append(str(BAK) + "-" + str(BBK))
    tepi_kelas.append(tepi)
tepi_kelas.append(BBK + 0.5)
print("6. Kelas =", kelas)
print("6. Interval kelas =", interval_kelas)
print("6. Tepi kelas =", tepi_kelas)


# 7. Menghitung frekuensi tiap kelas

# Convert data into a pandas DataFrame
df = pd.DataFrame(data, columns=["Value"])

# Create frequency distribution using pd.cut and value_counts
df["Range"] = pd.cut(df["Value"], bins=tepi_kelas, labels=interval_kelas, include_lowest=True)
frequency_table = df["Range"].value_counts().sort_index()
relative_frequency = frequency_table / frequency_table.sum() * 100

# Convert to DataFrame for a cleaner look
frequency_distribution = pd.DataFrame(
    {
        "Kelas": kelas,
        "Range": frequency_table.index,
        "Frequency": frequency_table.values,
        "Relative Frequency (%)": relative_frequency,
    }
)
frequency_distribution.reset_index(drop=True, inplace=True)


# Display the frequency distribution table
print("\nTable of Frequency Distribution")
print(frequency_distribution)
```

## 2. Histogram

Grafik yang menampilkan distribusi frekuensi dari suatu variabel numerik.

Langkah-Langkah:
1. Gunakan tabel distribusi frekuensi
2. Gunakan tepi kelas untuk titik absis pada sumbu x dan frekuensi absolut untuk titik absis pada sumbu y
3. Gambarlah balok dengan tinggi sesuai frekuensi absolut dan lebar sesuai dengan tepi kelas

```python
import matplotlib.pyplot as plt

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.bar(
    frequency_distribution["Range"],
    frequency_distribution["Frequency"],
    width=1,
    edgecolor="black",
    color="skyblue",
)
plt.title("Histogram of Frequency Distribution")
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.grid(axis="y")

# Show the plot
plt.tight_layout()
plt.show()
```
