## Case Study

Didalam sebuah ruangan, terdapat 12 orang yang sedang berkumpul dengan umur sebagai berikut:

25 | 12 | 20 | 20 | 15 | 21 | 22 | 19 | 19 | 24 | 17 | 65

Coba analisis pusat distribusi data, untuk melihat apakah ada outlier pada data tersebut.


## a. Range
**Rentang (range)** adalah ukuran statistik yang mengukur seberapa besar perbedaan antara nilai terbesar dan nilai terkecil dalam kumpulan data. Rentang dihitung dengan mengurangi nilai terbesar (maksimum) dan nilai terkecil (minimum) dalam kumpulan data.

````python
# Fungsi untuk menghitung rentang
def calculate_range(data):
    nilai_tertinggi = max(data)
    print(f"Nilai tertinggi: {nilai_tertinggi}")

    nilai_terendah = min(data)
    print(f"Nilai terendah: {nilai_terendah}")

    # Selisih antara nilai tertinggi dan terendah
    data_range = max(data) - min(data)
    
    print(f"\n")    
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")  
    
    return data_range


# Inisialisasi dataset
data = [25, 12, 20, 20, 15, 21, 22, 19, 19, 24, 17, 65]
print(f"Data: {data}")

data_range = calculate_range(data)
````

Jika divisualisasikan dalam bentuk grafik, maka Rentang (range) akan terlihat sebagai berikut:

````python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Define the horizontal line coordinates
y = 1.5 # The y-coordinate where the line will be drawn
x1 = 4/70*1  # The x-coordinate of the start point
x2 = 66/70*1  # The x-coordinate of the end point

# Add a horizontal line
plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {calculate_range(data)}')

# Add vertical line
plt.axvline(x=min(data), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(data):.2f}')
plt.axvline(x=max(data), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(data):.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range')

# Add legend
plt.legend()

# Show the plot
plt.show()
````

Dari grafik tersebut, kita dapat melihat bahwa rentang data sangat besar, yaitu antara **12 hingga 65**. Namun, rentang memiliki keterbatasan karena hanya mempertimbangkan dua nilai ekstrem dan tidak memperhitungkan distribusi nilai-nilai di antara keduanya.

Oleh karena itu, dalam beberapa kasus, statistik lain seperti **variansi atau standar deviasi** dapat memberikan gambaran yang lebih komprehensif tentang sebaran data.

## b. Variance
**Variansi** adalah ukuran statistik yang mengukur sebaran atau variasi data dari rata-rata. Variansi dihitung dengan menjumlahkan kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data. 

Kita dapat menghitung variansi secara manual di Python dengan mengikuti langkah-langkah berikut:

1. Hitung rata-rata dari dataset
2. Hitung selisih antara setiap titik data dengan rata-rata, kemudian kuadratkan hasilnya
3. Hitung rata-rata dari kuadrat selisih, dengan menjumlahkan kuadrat selisih antara setiap titik data dengan rata-rata, kemudian dibagi dengan jumlah total data

````python
# Fungsi untuk menghitung rata-rata
def calculate_mean(data):
    return sum(data) / len(data)

# Fungsi untuk menghitung variansi
def calculate_variance(data):
    # Rata-rata 
    mean = calculate_mean(data)
    print(f"Rata-rata: {mean}")  
    
    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")    
    squared_diff = []
    for x in data:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )
        print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")    
    
    print(f"Total kuadrat selisih: {sum(squared_diff)}")    
    print(f"Jumlah total data: {len(data)}")    
    
    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(data)
    
    print(f"\n")    
    print(f"Variansi: {sum(squared_diff)} / {len(data)} = {variance}")    
    
    return variance


# Inisialisasi dataset
data = [25, 12, 20, 20, 15, 21, 22, 19, 19, 24, 17, 65]
print(f"Data: {data}")

variance = calculate_variance(data)
````

Perhatikan bahwa nilai **varians** yang didapatkan sangat besar. Hal ini terjadi karena varians diukur dengan cara mengkuadratkan nilai-nilai tersebut.

Kita mungkin bertanya-tanya, apa manfaat dari nilai yang sangat besar ini?

Nah, sebenarnya ada konsep statistik lain yang lebih mudah dipahami ketika ingin mengetahui seberapa jauh data dari rata-rata, yaitu **standar deviasi**. Jadi, varians itu penting untuk menghitung standar deviasi, karena standar deviasi dapat memberikan gambaran yang lebih jelas tentang sebaran data.

## c. Standard Deviation

**Standar deviasi** adalah ukuran statistik yang memberikan gambaran tentang seberapa jauh data tersebar dari rata-rata. Semakin tinggi standar deviasi, semakin besar variasi atau sebaran data. Standar deviasi adalah akar kuadrat dari variansi.

Kita dapat menghitung standar deviasi secara manual di Python dengan mengikuti langkah-langkah berikut:

1. Hitung variansi atau rata-rata kuadrat selisih (poin b)
2. Hitung akar kuadrat dari variansi

````python
# Fungsi untuk menghitung standar deviasi
def calculate_standard_deviation(data):
    variance = calculate_variance(data)
    standard_deviation = variance**0.5
    print(f"\n")   
    print(f"Standard Deviation: {variance}**0.5 = {standard_deviation}")
    
    return standard_deviation

# Inisialisasi dataset
data = [25, 12, 20, 20, 15, 21, 22, 19, 19, 24, 17, 65]
print(f"Data: {data}")

standard_deviation = calculate_standard_deviation(data)
mean = calculate_mean(data)
````

Jika divisualisasikan dalam bentuk grafik, maka Standar Deviasi akan terlihat sebagai berikut:

````python
import matplotlib.pyplot as plt
import numpy as np

# Buat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Tambahkan garis vertikal pada standar deviasi
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-standard_deviation:.2f}')
plt.axvline(x=mean + standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean+standard_deviation:.2f}')

# Tambahkan label dan judul
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Std Deviation = {standard_deviation:.2f}')

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
````

Terlihat dari grafik bahwa kita memiliki standar deviasi yang sangat besar, yaitu sekitar 13 dari rata-rata (dalam rentang 10,20 - 36,30). Ini berarti bahwa data umur dalam dataset kita, secara rata-rata, berbeda sebesar 13 dari rata-rata. Penting untuk diingat bahwa ini adalah ukuran dispersi dan bukan selisih umur individual.

Semakin dekat nilai-nilai dalam dataset dari rata-rata, semakin kecil standar deviasi. Namun, dalam kasus kita, nilai dari standar deviasi terdistorsi jauh karena kita memiliki **outlier**, yaitu nilai 65.

Sekarang, untuk melihat bagaimana standar deviasi berubah, mari kita hilangkan data **outlier**, yaitu 65. Dataset umur kita sekarang hanya memiliki 11 nilai.

25 | 12 | 20 | 20 | 15 | 21 | 22 | 19 | 19 | 24 | 17

````python
# Inisialisasi dataset tanpa outlier
data = [25, 12, 20, 20, 15, 21, 22, 19, 19, 24, 17]
print(f"Data: {data}")

standard_deviation = calculate_standard_deviation(data)
mean = calculate_mean(data)
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Buat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Tambahkan garis vertikal pada standar deviasi
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-standard_deviation:.2f}')
plt.axvline(x=mean + standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean+standard_deviation:.2f}')

# Tambahkan label dan judul
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Std Deviation = {standard_deviation:.2f}')

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
````

Pada grafik yang baru, kita memiliki standar deviasi yang lebih kecil, yaitu sebesar 3,6 dari rata-rata (dalam rentang 15,85 - 23). Tanpa adanya data outlier, nilai dari standar deviasi turun secara dramatis, dari 13 menjadi 3. 

### Lebih baik Standar Deviasi yang kecil atau besar?

Standar deviasi yang lebih kecil menunjukkan bahwa data cenderung lebih rapat di sekitar rata-rata, sedangkan standar deviasi yang lebih besar menunjukkan bahwa data lebih tersebar atau memiliki variasi yang lebih besar dari rata-rata.

Dalam banyak kasus, memiliki standar deviasi yang kecil lebih baik karena menunjukkan tingkat konsistensi atau prediktabilitas yang lebih tinggi dalam dataset. Namun, ada situasi di mana standar deviasi yang lebih besar dapat memberikan informasi yang berharga, terutama jika variasi yang besar adalah hal yang diharapkan atau relevan dalam konteks tertentu.
